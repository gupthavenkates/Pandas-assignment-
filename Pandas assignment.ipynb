```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
df
```

```python
#1.The missing numbers are increased by 10 in flight number column and type is int
df.fillna("Fill",downcast='infer',inplace=True)
temp=np.array(df.FlightNumber)
for num in range(len(temp)):
    if temp[num] == "Fill":
        temp[num] = temp[num-1]+10
df.FlightNumber=temp
df
```

```python
#spliting from_to into two columns and capitalizing each column
x=(df["From_To"].str.split('_',expand = True))
From= x[0].str.capitalize()
To = x[1].str.capitalize()
print(From,To)
```

```python
#inserting 'From' & 'To' column in the dataframe at correct location
df.insert(0,"From",From)
df.insert(1,"To",To)
df
```

```python
#Removing From_To column from dataframe
df.drop(columns="From_To",inplace=True)
df
```

```python
#Creating delay_1,delay_2,delay_3
delays=df['RecentDelays']
delays=pd.DataFrame(delays.tolist(),columns=('delay_1','delay_2','delay_3'))
delays
```

```python
#Replaceing RecentDelays with delay_1,delay_2,delay_3
df.insert(3,'delay_1',delays['delay_1'])
df.insert(4,'delay_2',delays['delay_2'])
df.insert(5,'delay_3',delays['delay_3'])
df.drop(columns='RecentDelays',inplace=True)
df
```
